```python
# Problem 5 : Leap year
"""
year % 4 == 0 T
and
year % 100 == 0 F
or
year % 400 == 0 T
"""

def is_leap(year):
    
    if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
        return True
    else:
        return False

year = int(input('year : '))
print(is_leap(year))
```
