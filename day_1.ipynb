### Variables and datatypes

```python
name = 'chiru'
age = 24
gpa = 8.85
is_studying = False
job = None

print(type(name))
print(type(age))
print(type(gpa))
print(type(is_studying))
print(type(job))
```

### String methods

```python
sentence = "Now is better than never"

print(sentence.strip()) #remove spaces

print(sentence.upper()) 

print("Title : ",sentence.title())

print(sentence.replace('Now', 'Today'))

print(len(sentence))

print('better' in sentence)

words = sentence.strip().split(" ")
print(words)
print("-".join(words))
```

### f-string and type conversion

```python
name = 'chiru'
age = 24
gpa = 8.85

#f string
print(f"my name is {name},age {age},GPA {gpa:.1f}")

#Type conversion
print(int("43"))
print(float("345.23"))
print(int(23.345))
print(str(59))
print(bool(0))
print(bool(1))

#arithmetic

print(17//5)
print(17%5)
print(2 ** 10)
```

```python
# Radius of circle
pi = 3.14159
diameter = 3

radius = diameter/2

area = pi * radius ** 2
print(f'area of cirle with diameter of {diameter} is {area}')
```

```python
# value swap
a = 23
b = 56

print(f'initial value of a is {a} and b is {b}')
temp = a
a = b
b = temp
print(f'after swapping value of a is {a} and b is {b}')

"""
a, b = b, a
"""
```
