## Word search 

```python
doc_list = ["The Learn Python Challenge Casino.", "The bought a car", "The Casinoville"]
keyword = 'casino'
```

```python
def word_search(doc, keyword):
    indices = []
    for i, doc in enumerate(doc_list):
        tokens = doc.split()
        #print(tokens)
        normalized = [token.rstrip('.,').lower() for token in tokens]
        #print(normalized)
        if keyword in normalized:
            indices.append(i)
    
    return indices

print(word_search(doc_list,keyword))
```

### Multi word search

```python
doc2_list = ["The Learn Python Challenge Casino.", "They bought a car and a casino", "Casinoville"]
m_keywords = ['casino','they']
k_dict = {}
```

```python
def multiword_search(documents, keywords):
    key_dict = {}
    for key in m_keywords:
        key_dict[key] = word_search(documents, key)
    return key_dict

print(multiword_search(doc2_list,m_keywords))
```
