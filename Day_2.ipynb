List

```python
fruits = ["apple", "banana", "mango", "orange"]

print(fruits[0]) #indexing
print(fruits[-1])
print(fruits[1:3]) #slicing
```

```python
fruits.append("grape")
```

```python
print(fruits)
```

```python
fruits.insert(1,"watermelon")
```

```python
print(fruits)
```

```python
fruits.remove('banana')
print(fruits)
```

```python
popped = fruits.pop()
print(popped)
```

```python
print(len(fruits))
```

```python
print(sorted(fruits)) #sorted doesnot affect the original list
```

```python
print(fruits)
```

List comprehension

```python
squares = [x**2 for x in range(1,6)]
print(squares)
```

```python
evens = [x for x in range(20) if x%2==0]
print(evens)
```
