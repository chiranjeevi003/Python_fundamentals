### 1. File Reading and writing

```python
#writing to a file
with open("scores.txt", "w") as f:
    f.write("chiru,85\n")
    f.write("dusky,92\n")
    f.write("hippo,67\n")
    f.write("arjun,78\n")
    
#Reading the entire file
with open("scores.txt","r") as f:
    content = f.read()
    print(content)
```

```python
#Reading line by line (memory efficient - important for big file)

with open("scores.txt","r") as f:
    for line in f:
        line = line.strip()
        name, score = line.split(",")
        print(f'{name} scored {score}')
        
```

```python
#appending
with open("scores.txt",'a') as f:
    f.write("deepak,55\n")
```

```python
#reading into a list of lines
with open("scores.txt",'r') as f:
    lines = f.readlines()
    
print(lines)
```

### 2. Working with CSV 

```python
#write csv
import csv
rows = [
    ["name", "scores", "grade"],
    ["chiru", "85","B"],
    ["dusky", "92", "A"],
    ["hippo", "67", "C"],
]
```

```python
with open("students.csv","w",newline="") as f:
    writer = csv.writer(f)
    writer.writerows(rows)
```

```python
# read csv
with open("students.csv","r") as f:
    reader = csv.DictReader(f)
    
    for row in reader:
        print(dict(row))
```

```python
#Read CSV manually 
with open('students.csv','r') as f:
    lines = f.readlines()
    header = lines[0].strip().split(',')
    for line in lines[1:] :
        values = line.strip().split(',')
        record = dict(zip(header,values))
        print(record)
```

### 3. Error handling

```python
try : 
    x = int(input("enter a number : "))
    result = 100 / x
    print(f"100 / {x} = {result}")
    
except ValueError:
    print("Thats not a valid number")
    
except ZeroDivisionError:
    print("Cannot divide by Zero !")
    
except Exception as e:
    print(f"Unexpected Error : {e}")
    
finally:
    print("This always runs - use for cleanup")
    
    
```

```python
#Raising exceptions

def get_score(score):
    if not isinstance(score, (int, float)):
        raise TypeError(f'Score must be a number, got {type(score)}')
    if not 0 <= score <= 100:
        raise ValueError(f"Score must be 0-100, got{score}")
        
    return score

try:
    get_score(150)
except ValueError as e:
    print(f"Validation Error : {e}")
```
