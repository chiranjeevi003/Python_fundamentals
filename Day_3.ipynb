### functions

```python
def greet(name, greeting='Hello'):
    return f'{greeting}, {name}'

print(greet('Chiru'))
print(greet('Priya', 'Hi'))

def add_all(*numbers):
    return sum(numbers)

print(add_all(1,2,3,4,5))

# **kwargs - variable keyword argument

def describe(**info):
    for key, val in info.items():
        print(f'{key} : {val}')
        
describe(name='chiru', city='Tumakur', skill='python')


#function returning multiple value
def min_max(numbers):
    return min(numbers), max(numbers)

lo, hi = min_max([4,7,9,2,1])
print(f'Min = {lo}, max = {hi}')


#lambda functions

square = lambda x:x**2
double = lambda x:x*2
add = lambda x, y : x+y

print(square(5))
print(double(8))
print(add(59, 34))

#lambda with sorted / map/ filters 

students = [('chiru', 89), ('dusky', 93),('hehe',67)]
students_sorted = sorted(students, key = lambda s:s[1], reverse=True)
print(students_sorted)

nums = [1,2,3,4,5,6]
double = list(map(lambda x : x*2, nums))
evens = list(map(lambda x: x%2==0, nums))

print(double)
print(evens)
```

### sets

```python
a = {1,2,3,4,5}
b = {3,4,5,6,7}

print(a | b) #union
print(a & b) #intersection
print(a - b) #only a or differnece
print(a ^ b) #symetric difference
```

```python
#removing duplicates

raw = [1,2,2,3,3,3,4]
unique = list(set(raw))
print(unique)
```
